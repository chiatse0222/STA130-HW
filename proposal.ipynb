# The Role of Social Anxiety in Shaping Social Behaviors, Preferences, and Support Reliance

## Analysis 1: Social Anxiety and Frequency of Social Interactions across Various Contexts

### Research Question
How does social anxiety influence the frequency of meaningful social interactions across different contexts?

### Variables and Justification

- **Variables**:
    - `CONNECTION_meaningful_interactions`: Frequency of meaningful social interactions.
    - `CONNECTION_social_days_family_p7d_grouped`: Days spent with family in the past week.
    - `CONNECTION_social_days_friends_p7d_grouped`: Days spent with friends in the past week.
    - `CONNECTION_social_days_coworkers_and_classmates_p7d`: Days spent with coworkers/classmates in the past week.
    - `PSYCH_social_phobia_scale_walking`: Anxiety related to walking in public.
    - `PSYCH_social_phobia_scale_attention`: Anxiety about being the center of attention.
    - `PSYCH_social_phobia_scale_eating`: Anxiety related to eating in public.
    - `PSYCH_social_phobia_scale_conspicuous`: Anxiety about appearing conspicuous.

- **Justification**: These variables capture the frequency and context of social interactions while also including specific social anxiety factors tied to public settings. Together, they provide a comprehensive view of how social anxiety may reduce engagement in various social contexts.

### Visualizations

- **Histograms** for each type of social engagement, showing distribution patterns across contexts.
- **Box plots** for meaningful interactions, comparing levels of engagement across different levels of social anxiety.

### Planned Analysis

- **Sampling Distributions and Confidence Intervals**: Bootstrapping will generate sampling distributions and calculate 95% confidence intervals for each regression coefficient.
- **Purpose**: To quantify the effect of social anxiety on social engagement and assess whether higher levels of anxiety correlate with reduced interaction in specific contexts.

### Hypothesis
Individuals with higher social anxiety levels are expected to engage less frequently in meaningful interactions and spend less time with family, friends, and coworkers. This would suggest that targeted interventions promoting social engagement could benefit individuals with high social anxiety.

## Analysis 2: Social Anxiety and Preferences for Social versus Solitary Activities

### Research Question
How does social anxiety influence preferences for social versus solitary activities, particularly in exercise and recreational settings?

### Variables and Justification

- **Variables**:
    - `EXERCISE_social_more_likely_to_engage`: Likelihood to engage in exercise with others.
    - `EXERCISE_social_not_likely_when_alone`: Likelihood to avoid exercise when alone.
    - `EXERCISE_social_enjoy_team_sports_more`: Preference for team sports.

- **Justification**: These variables assess preferences for social versus solitary activities, specifically around exercise and recreation. Understanding these preferences provides insights into how social anxiety might discourage participation in group activities.

### Visualizations

- **Bar charts** for preferences related to social exercise, showing differences between individuals with high and low social anxiety.
- **Box plots** for team sports enjoyment, illustrating how social anxiety influences preference for group versus solo settings.

### Planned Analysis

- **Bootstrapped Confidence Intervals**: Bootstrapping will generate confidence intervals for regression coefficients predicting exercise engagement preferences and enjoyment of social activities.
- **Purpose**: To confirm the effect of social anxiety on preferences for solitary versus social activities, and to understand if high social anxiety correlates with avoidance of group settings.

### Hypothesis
Individuals with higher social anxiety are expected to prefer solitary activities and may have reduced enjoyment of team sports. This would indicate the need for programs that encourage comfortable social engagement in physical activities for individuals with high social anxiety.

## Analysis 3: Predicting Social Isolation and Support Needs Based on Social Anxiety and Demographics

### Research Question
What demographic and psychological factors predict feelings of social isolation and reliance on support systems?

### Variables and Justification

- **Variables**:
    - `DEMO_age`, `DEMO_gender`, `GEO_province`: Demographic variables to explore differences in isolation patterns across age, gender, and region.
    - `DEMO_identity_disability`: Disability identity, which may influence social support needs.
    - `CONNECTION_social_days_neighbours_p7d_grouped`: Days spent interacting with neighbors, representing social engagement outside immediate family and friends.
    - `CONNECTION_social_num_close_friends_grouped`: Number of close friends, indicating the size of an individual's support network.
    - `ADVICE_social_skills`: Guidance on which social skills individuals consider important to develop for better social interaction, potentially reflecting perceived support needs related to social anxiety.
    - `ADVICE_building_friends`: Advice on strategies for building friendships, which could correlate with efforts to reduce social isolation.
    - `ADVICE_barriers`: Common barriers to social connection, which provide insight into factors contributing to isolation.

- **Justification**: These demographic and social factors likely contribute to levels of social isolation and support reliance. Disability status, close friends, and neighborhood interactions help identify the support networks available to individuals, while advice on social skills and barriers to social connection offer insight into areas where people feel isolated or require support.

### Visualizations

- **Scatter plots** for continuous variables, such as age and isolation levels.
- **Histograms** for categorical demographic variables, showing isolation distribution across groups.

### Planned Analysis

- **Residual Analysis and Bootstrapping**: Residual analysis will confirm model assumptions, and bootstrapping will generate robust confidence intervals for regression coefficients.
- **Purpose**: To ensure reliable predictions and identify key factors associated with social isolation and support reliance.

### Hypothesis
Social anxiety, combined with demographic factors (such as age, province, and number of close friends), will predict social isolation and reliance on support. This analysis can guide targeted interventions for individuals who may feel isolated or have higher support needs due to social anxiety.
